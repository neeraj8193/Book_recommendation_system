```python
rating_with_name['Book-Rating'] = pd.to_numeric(rating_with_name['Book-Rating'], errors='coerce')
```

```python
avg_rating_df=rating_with_name.groupby('Book-Title')
```

```python
# prompt: calculatethe avg  rating of num

avg_rating_df['avg_rating'] = avg_rating_df['Book-Rating'].mean()
```

```python
avg_rating_df = rating_with_name.groupby('Book-Title')['Book-Rating'].agg(lambda x: x.astype(float).mean()).reset_index()
avg_rating_df.rename(columns = {'Book-Rating' : 'avg_rating'}, inplace = True)
avg_rating_df
```

```python
popularity_df=num_rating_df.merge(avg_rating_df,on='Book-Title')
```

```python
#collaborative filtering based recommender system
```

```python
x=rating_with_name.groupby("User-ID").count()['Book-Rating']>50
```

```python
padhe_likhe_users=x[x].index
```

```python
filter_rating=rating_with_name[rating_with_name['User-ID'].isin(padhe_likhe_users)]
```

```python
y=filter_rating.groupby('Book-Title').count()['Book-Rating']>10
```

```python
famous_book=y[y].index
```

```python
famous_book
```

```python
final_ratings=filter_rating[filter_rating['Book-Title'].isin(famous_book)]
```

```python
final_ratings.drop_duplicates()
```

```python
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.shape
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pt)
```

```python
def recommend(book_name):
  #index_fetch
  index=np.where(pt.index==book_name)[0][0]
  similar_items=sorted(list(enumerate(similarity_score[index])), key=lambda x:x[1],reverse=True)[1:9]
  for i in similar_items:
    print(pt.index[i[0]])

```
